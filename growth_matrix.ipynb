```python
#Based on Boston Consulting Group growth matrix
#organizes URL by click data from Google Analytics/Search Console
#to track growth over a 3 month period
#then organizes into 4 different bins cash cows, stars, question marks, dogs
#using sample data from google data studio
```

```python
import pandas as pd
import datetime
import numpy as np
import seaborn as sns
```

```python
e = input('Enter Your 3 Months in ascending Order (ex Jun 2022, Jul 2022)')
months = e.split(', ')
months.insert(0, 'url')
print(months)

#allows input of months in accepted format for working with larger data set to choose 3 month interval
```

```python
df = pd.read_csv('sample_data.csv')
df.rename(columns={'Landing Page': 'url', 'Date': 'date', 'Url Clicks': 'clicks'}, inplace=True)
#import api gsc data from datastudio 


dff = df.groupby(['url', 'date'])[['clicks']].agg('sum', 'mean')
#group articles by url & include date to make sure no duplicates, agrgrate sum & mean

df2 = pd.DataFrame(dff)
df2.sort_values('clicks', ascending=True, inplace=True)
df2.reset_index(inplace=True)

df3 = df2.pivot(index='url', columns='date', values='clicks')
#pivot so it is organized how we want, URL, Date, Clicks


df3.reset_index(inplace=True)
columns = list(df3.columns.values)

df3.sort_values(columns[3], ascending=False, inplace=True)



df3 = df3[months]

df3

```

```python
def foo(pct1, pct2, pct3):
    growth_over1 = (pct2 - pct1)/(pct1)
    growth_over2 = (pct3 - pct2)/(pct2)

    return growth_over2, growth_over1

def growth(m1, m2):
    s1 = ((m1-m2)/m2)
    return s1
```

```python

#add in 3 months total for the dataframes


cow = []
cow_value = []
cow_value2 = []
cow_value1 = []
cow_growth_pct = []
c_avg = []

star = []
star_value= []
star_value2 = []
star_value1 = []
star_growth_pct = []
s_avg = []

question = []
question_value = []
q_value2 = []
q_value1 = []
question_growth_pct = []
q_avg = []


dog = []
dog_value = []
dog_value1 = []
dog_value2 = []
dog_growth_pct = []
d_avg = []



null = []
mnth3 = df3[months[3]]
mnth2 = df3[months[2]]
mnth1 = df3[months[1]]




for i in range(0, len(df3['url'])):
    if ((((((mnth3[i] - mnth2[i])/mnth2[i]) < 0.1) and (((mnth2[i] - mnth1[i])/mnth1[i]) < 0.1))
        or (((mnth3[i]-mnth1[i])/mnth1[i])/2)<0.15)) and (mnth3[i] > 50) and (mnth2[i] > 50):
        if((growth(mnth3[i], mnth2[i]) > 0) or (growth(mnth2[i], mnth1[i]) > 0)):
            cow.append(df3['url'][i])
            cow_value.append(mnth3[i])
            cow_value1.append(mnth1[i])
            cow_value2.append(mnth2[i])
            cow_growth_pct.append(foo(mnth1[i], mnth2[i], mnth3[i]))
            c_avg.append((mnth1[i] + mnth2[i] + mnth3[i])/3)
    #checks if growth is below 10% or average growth is less than 15% and clicks above 7000 for 3 months-> cow
    if (((((mnth3[i] - mnth2[i])/mnth2[i])) > 0.1) and ((mnth2[i] - mnth1[i])/mnth1[i] > 0.1) or (((mnth3[i]-mnth1[i])/mnth1[i])/2)>0.15) and (mnth3[i] > 30) and (mnth2[i] > 30):
       star.append(df3['url'][i])
       star_value.append(mnth3[i])
       star_value1.append(mnth1[i])
       star_value2.append(mnth2[i])
       star_growth_pct.append(foo(mnth1[i], mnth2[i], mnth3[i]))

        #checks if growth is above 10%  and clicks above 5000 ->  star

    if ((((mnth3[i] - mnth2[i])/mnth2[i])) > 0.2) and ((mnth2[i] - mnth1[i])/mnth1[i] > 0.2) and (15 < mnth3[i] < 25):
       question.append(df3['url'][i])
       question_value.append(mnth3[i])
       q_value1.append(mnth1[i])
       q_value2.append(mnth2[i])
       question_growth_pct.append(foo(mnth1[i], mnth2[i], mnth3[i]))



       #checks if growth is above 20% and clicks below 800 but greater than 200 -> ?
    
    if ((((mnth3[i] - mnth2[i])/mnth2[i])) < 0.05) and ((mnth2[i] - mnth1[i])/mnth1[i] <0.05) and (mnth3[i] < 10):
        dog.append(df3['url'][i])
        dog_value.append(mnth3[i])
        dog_value1.append(mnth1[i])
        dog_value2.append(mnth2[i])
        dog_growth_pct.append(foo(mnth1[i], mnth2[i], mnth3[i]))



        
        #checks if growth is below 5% and clicks below 500 -> dog
    else: 
        null.append(df3['url'][i])


#for loop to categorize data into correct bins
```

```python
cash_cows = pd.DataFrame()
cols = ['%s_users' % months[1], '%s_users' % months[2], '%s_users' % months[3]]

cash_cows.insert(0, 'cash_cows', cow)
cash_cows.insert(1, '%s_users' % months[3], cow_value)
cash_cows.insert(2, '%s_users' % months[2], cow_value2)
cash_cows.insert(3, '%s_users' % months[1], cow_value1)
cash_cows[cols] = cash_cows[cols].astype('float64')
cash_cows.insert(4, 'cash_cows.avg', cash_cows[cols].mean(axis=1))
cash_cows.insert(2, 'growth', cow_growth_pct)
cash_cows[['%s_growth' % months[3], '%s_growth' % months[2]]] = (cash_cows['growth'].to_list())
cash_cows.drop(['growth'], axis=1, inplace=True)
cash_cows.iloc[:,1]=cash_cows.iloc[:,1].apply('{:,}'.format)
cash_cows.iloc[:,2]=cash_cows.iloc[:,2].apply('{:,}'.format)
cash_cows.iloc[:,3]=cash_cows.iloc[:,3].apply('{:,}'.format)
cash_cows.iloc[:,4]=cash_cows.iloc[:,4].apply('{:,}'.format)
cash_cows['%s_growth' % months[2]] = cash_cows['%s_growth' % months[2]].apply("{0:.4%}".format)
cash_cows['%s_growth' % months[3]] = cash_cows['%s_growth' % months[3]].apply("{0:.4%}".format)
cash_cows.set_index('cash_cows', inplace=True)
cash_cows.sort_values(('%s_growth' % months[3]), ascending=False, inplace=True)




stars = pd.DataFrame()
stars.insert(0, 'stars', star)
stars.insert(1, '%s_users' % months[3], star_value)
stars.insert(2, '%s_users' % months[2], star_value2)
stars.insert(3, '%s_userss' % months[1], star_value1)
stars.insert(2, 'growth', star_growth_pct)
stars[['%s_growth' % months[3], '%s_growth' % months[2]]] = (stars['growth'].to_list())
stars.drop(['growth'], axis=1, inplace=True)
stars.sort_values(('%s_growth' % months[3]), ascending=False, inplace=True)

stars['%s_growth' % months[3]] = stars['%s_growth' % months[3]].apply("{0:.4%}".format)
stars['%s_growth' % months[2]] = stars['%s_growth' % months[2]].apply("{0:.4%}".format)
stars.iloc[:,1]=stars.iloc[:,1].apply('{:,}'.format)
stars.iloc[:,2]=stars.iloc[:,2].apply('{:,}'.format)
stars.iloc[:,3]=stars.iloc[:,3].apply('{:,}'.format)
stars.set_index('stars', inplace=True)




dogs = pd.DataFrame()
dogs.insert(0, 'dogs', dog)
dogs.insert(1, '%s_users' % months[3], dog_value)
dogs.insert(2, '%s_users' % months[2], dog_value2)
dogs.insert(3, '%s_users' % months[1], dog_value1)
dogs.insert(2, 'growth',dog_growth_pct)
dogs[['%s_growth' % months[3], '%s_growth' % months[2]]] = (dogs['growth'].to_list())
dogs.drop(['growth'], axis=1, inplace=True)
dogs['%s_growth' % months[3]] = dogs['%s_growth' % months[3]].apply("{0:.4%}".format)
dogs['%s_growth' % months[2]] = dogs['%s_growth' % months[2]].apply("{0:.4%}".format)
dogs.iloc[:,1]=dogs.iloc[:,1].apply('{:,}'.format)
dogs.iloc[:,2]=dogs.iloc[:,2].apply('{:,}'.format)
dogs.iloc[:,3]=dogs.iloc[:,3].apply('{:,}'.format)
dogs.sort_values(('%s_growth' % months[3]), ascending=True, inplace=True)
dogs.set_index('dogs', inplace=True)



question_marks = pd.DataFrame()
question_marks.insert(0, 'question_marks', question)
question_marks.insert(1, '%s_users' % months[3], question_value)
question_marks.insert(2, '%s_users' % months[2], q_value2)
question_marks.insert(3, '%s_users' % months[1], q_value1)
question_marks.insert(2, 'growth', question_growth_pct)
question_marks[['%s_growth' % months[3], '%s_growth' % months[2]]] = (question_marks['growth'].to_list())
question_marks.drop(['growth'], axis=1, inplace=True)
question_marks.iloc[:, 4] = question_marks.iloc[:, 4].apply("{0:.4%}".format)
question_marks.iloc[:, 5] = question_marks.iloc[:, 5].apply("{0:.4%}".format)
question_marks.sort_values(('%s_growth' % months[3]), ascending=False, inplace=True)

question_marks.set_index('question_marks',inplace=True)


#insert data into dataframes
```
