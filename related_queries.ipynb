```python

#import the libraries
import pandas as pd                        
from pytrends.request import TrendReq


#build model
pytrend = TrendReq(hl='en-US')

#provide your search terms
words = input("enter keywords seperated by comma: ")
kw_list = words.split(', ')


pytrend.build_payload(kw_list=kw_list, timeframe = 'now 7-d', geo = "US")

df1 = pd.DataFrame()





```

```python
print("\nleft columns are top & right columns are rising queries ")

for x in range(len(kw_list)):
        related_queries = pytrend.related_queries()
        related_queries.values()
       
        top = list(related_queries.values())[x]['top']
        rising = list(related_queries.values())[x]['rising']
        
        dftop = pd.DataFrame(top)
        dfrising = pd.DataFrame(rising)
        
        joindfs = [dftop, dfrising]
        allqueries = pd.concat(joindfs, axis=1)

        
        
        allqueries.rename({'query': 'top query', 'value': 'top query value', 'query.1': 'top query', 'value.1': 'related query value'}, axis=1, inplace=True) 
        
        print("\n", kw_list[x], "related queries: ")
        display(allqueries.head(50))
        df1 = df1.append(allqueries)
        df1.reset_index(inplace=True, drop=True)

        
    
        
        x+=1
        
        



```

```python
df1.to_csv('capital1_trends.csv')
```
